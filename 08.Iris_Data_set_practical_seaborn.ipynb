```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
iris.head()
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
# Continuous data set
# joint plot
sns.jointplot(x=iris['sepal_length'],y=iris['petal_length'],kind='reg')
```

```python
# Dist plot
sns.displot(iris)
```

```python
sns.distplot(iris['petal_width'])
```

```python
# pair plot
sns.pairplot(iris)
plt.savefig('Image/pairPlot_Iris.png')
```

```python
# pair plot
sns.pairplot(iris,hue="species")
plt.savefig('Image/pairPlot_Iris_species.png')
```

```python
# Categorial Data set analysis
```

```python
#count plot
sns.countplot(iris['petal_width'])
```

```python
# bar plot
sns.barplot(x=iris['petal_length'],y=iris['petal_width'])
sns.set(rc={'figure.figsize':(19.7,6.27)})
```

```python
# box plot
sns.boxplot(x=iris['petal_length'],y=iris['petal_width'])
```

```python
# vilion plot
sns.violinplot(x=iris['petal_length'],y=iris['petal_width'])
```

```python
# vilion plot
sns.violinplot(x=iris['petal_length'],y=iris['petal_width'],hue=iris['species'])
```
