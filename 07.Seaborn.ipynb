## Contninuous Plot
- It is used to do data analysis of continuous data.
- Joint plot
- Dist plot
- Pair plot

-----------------------------------------------------------------------
### Joint Plot
- Univarite Plot
- Joint plot are performed by univariate analysis.
- It is used to find the relationsip between two numerical values.

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='hex') #hex - hexagonal
```

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='scatter')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='reg') # reg - regression
```

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='hist') # hist - histogram
```

### Pair Plot
- defien relationship between more than two numerical values.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='day')
```

```python
df['smoker'].value_counts()    # to count the values.
```

```python
df['sex'].value_counts()
```

```python
plt.pie(df['sex'].value_counts(),labels=df["sex"].value_counts())
plt.show()
```

### Dist plot
- It is used to check the distribution of the features.

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['total_bill'],kde=False)
```

## categorical Plot of iris data set
