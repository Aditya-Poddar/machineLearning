#### Import StringIO library to read CSV from string using StringIO

```python
from io import StringIO
```

```python
data=('index,c1,c2,c3\n'
      '0,4,8,12\n'
      '1,3,6,9\n'
      '2,5,10,15'
     )
```

```python
data
```

```python
df =pd.read_csv(StringIO(data))
```

```python
df
```

```python
df=pd.read_csv(StringIO(data),index_col=0)
```

```python
df
```

```python
data=('a,b,c\n'
     '4,apple,orange\n'
     '8,book,pen')
```

```python
df=pd.read_csv(StringIO(data))
```
