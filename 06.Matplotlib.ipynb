### Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.subplot(2,2,3)
plt.plot(x,y,'g*')
plt.subplot(2,2,4)
plt.plot(x,y,'y-.o')
plt.savefig('Image/subplot.png')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title("Sin Wave")
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sin Wave Form')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("Cos Wave Form")

plt.tight_layout()  # to add automatically proper space between images of subplot.
plt.savefig('Image/subplot_of_sin_cos.png')
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
y_tan=np.tan(x)

plt.subplot(2,3,1)
plt.plot(x,y_sin)
plt.title('Sin Wave Form')

plt.subplot(2,3,2)
plt.plot(x,y_cos)
plt.title("Cos Wave Form")

plt.subplot(2,3,3)
plt.plot(x,y_tan)
plt.title("Tan Wave Form")

plt.subplot(2,3,4)
plt.plot(y_tan,x)
plt.title("")


plt.subplot(2,3,5)
plt.plot(y_sin,x)
plt.title("")

plt.subplot(2,3,6)
plt.plot(y_cos,x)
plt.title("")

plt.tight_layout()  # to add automatically proper space between images of subplot.
plt.savefig('Image/subplot_of_sin_cos_tan.png')
```

### Bar Plot

```python
x=[2,3,4,5]
y=[1,2,8,9]

x2=[10,2,8,9]
y2=[11,7,4,12]

plt.bar(x,y,color='g')
plt.bar(x2,y2,color='r')
plt.xlabel("x axis")
plt.ylabel('y axis')
plt.title("Bar Plot ")

plt.savefig('Image/barplot.png')
```

### Histogram

```python
x=[22,2,3,11,23,12,7,33,87,65,54,45,56,98,78,89]

plt.hist(x,bins=10)
plt.xlabel("values")
plt.ylabel('Values density')
plt.title('Histogram')
plt.savefig('Image/histogram.png')
```

##### 2D histogram

```python
# 2D histogram
x=[22,2,3,11,23,12,7,33,87,65,54,45,56,98,78,89]
y = np.arange(1,17)

plt.hist2d(x,y)
plt.show()
plt.savefig('Image/2dHistogram.png')
```
