```python
import numpy as np 
import pandas as pd
```

```python
x=np.arange(0,11)
y=np.arange(10,21)
```

### Scatter Plot

```python
# draw scatter plot
plt.scatter(x,y)
```

- plt.scatter(x,y,c='g')   
- here c='g' c stands for colour 
- and g for green 
- b for blue
- r for red 
- y for yelow
- k for black

### Bar Plot

```python
plt.bar(x,y)
plt.xlabel("X axis")
plt.ylabel('y axis')
plt.title("Bar Graph")
plt.savefig("bar1.jpg") # to save the image
```

### Pie Chart

```python
plt.pie(x,labels=x)
plt.show()
```

```python
plt.title("pie chart")
plt.pie(x,labels=x)
plt.show()
```
